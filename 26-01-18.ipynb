Part 0: The dataset
The file bdastudents.csv (separator ;) contains a free variation of the dataset
"Student Performance" dataset available in the UCI Machine Learning Repository.
http://archive.ics.uci.edu/ml/datasets/Student+Performance
The file represents some data on students attending 2 courses in 2 different schools
and the field G3 the final evaluation (to be predicted).
The schema of the dataset is
School - school of the student (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da
Silveira)

separated)
